<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab: 
1. Perbedaan Series dengan List dan Dict :
- Series dgn List : Series dapat diubah atau di beri label pada index nya. misal index awal 0 1 2 3 menjadi a b c d sedangkan list tidak
- Series dgn Dict : Series dengan Dict sangat mirip namun dapat dibedakan dengan cara mengakses data nya dan tujuan pengunaan nya, Series biasanya menggunakan Label/Key yang berurutan seperti 0 1 2 3 atau a b c d, namun Dict memiliki Key yang tidak berurutan

2. Perbedaan Dataframe dan Series
DataFrame adalah tipe data 2 dimensi, sedangkan Series adalah tipe data 1 dimensi. DataFrame seperti tabel yang bisa menyimpan data dengan tipe yang berbeda, sedangkan Series seperti garis yang hanya bisa menyimpan data dengan tipe yang sama.

```python
# code here
p = pd.Series(person)
print(f'Series : \n{p}')
print(f"\nAccess 'Rangga Ariq' value : {p.fullname[1]}")
```
